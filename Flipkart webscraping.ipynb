```python
from autoscraper import AutoScraper
```

```python
url= "https://www.flipkart.com/q/bluetooth-headphones"

wanted_list = ["₹1,299 ","Boult Audio Z40 with Zen ENC Mic, 60H Battery Life, Low...","(43,485)",]
```

```python
scraper = AutoScraper()
result = scraper.build(url, wanted_list)
print(result)
```

```python
scraper.get_result_similar(url,grouped=True)
```

```python
scraper.set_rule_aliases({'rule_lmq5':'Title','rule_j4gl': 'Price'})
scraper.keep_rules(['rule_lmq5','rule_j4gl'])
scraper.save('flipkart-search')
```

```python
results = scraper.get_result_similar('https://www.flipkart.com/search?q=boat&otracker=search&otracker1=search&marketplace=FLIPKART&as-show=on&as=off',group_by_alias=True)
```
